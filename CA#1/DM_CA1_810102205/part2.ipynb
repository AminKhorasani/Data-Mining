import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pylab import *
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pylab as py
```

```python
# read data
df = pd.read_csv('ab.csv')
```

```python
# ascending datasets
a_s = df.sort_values(by='a', ascending=True)
b_s = df.sort_values(by='b', ascending=True)
```

```python
# split to 2 datset
a = a_s.filter(regex='a')
b = b_s.filter(regex='b')
```

```python
# calculating mean
a_mean = a.mean()
b_mean = b.mean()

# calculating std
a_std = a.std()
b_std = b.std()
```

```python
# finding Q1 and Q3
q1_a = np.percentile(a, 25)
q3_a = np.percentile(a, 75)
q1_b = np.percentile(b, 25)
q3_b = np.percentile(b, 75)
```

```python
print(q1_a, q1_b, q3_a, q3_b)
print(a_mean, b_mean)
print(a_std, b_std)
```

```python
# plotting box plot
df.plot.box(title='Boxplot of two dataset');
a.plot.box(title='Boxplot of A');
b.plot.box(title='Boxplot of B');
```

```python
#plotting histogram
sns.histplot(data=a, x="a").set_title('Histogram of A')
```

```python
# plotting histogram
sns.histplot(data=b, x="b").set_title('Histogram of B')
```

```python
# z-score normalization
a_zscore = a.apply(stats.zscore)
b_zscore = b.apply(stats.zscore)
```

```python
# plotting Q-Q plot
plt.scatter(a_zscore,b_zscore)
plt.plot([-2, 2],[-2,2],color="red")
plt.xlabel("A normalized dataset")
plt.ylabel("B normalized dataset")
plt.show()
```

```python
a1 = stats.zscore(a['a'])
b1 = stats.zscore(b['b'])

plt.figure(figsize=(8, 6))
stats.probplot(a1, dist="norm", plot=plt)
plt.show()

plt.figure(figsize=(8, 6))
stats.probplot(b1, dist="norm", plot=plt)
plt.show()
```

```python
# calculating correlation
correlation = a['a'].corr(b['b'])
```
