# Libraries

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

# Pre-processing

```python
df = pd.read_csv('weatherAUS.csv')
```

```python
df
```

```python
# round values of columns containing 'Temp'
temp_columns = [col for col in df.columns if 'Temp' in col]
df[temp_columns] = df[temp_columns].round(0)
```

```python
df.dtypes
```

```python
df.info()
```

```python
# change some colunn type to categorical value
df = df.astype({"Location":'category', "WindGustDir":'category',
                "WindDir9am":'category', "WindDir3pm":'category',
                "RainToday":'category', "RainTomorrow":'category'})
```

```python
df.info()
```

```python
df.dtypes
```

```python
# count NaN values of each column
nan_counts = df.isna().sum()
```

```python
nan_counts
```

```python
# handle NaN value

for column in df.columns:
    if nan_counts[column] < 10000:
        # drop rows containing NaN values
        df = df.dropna(subset=[column])

        # fill NaN values with mode of present values in the column

    else:
        try:
            df[column] = df[column].fillna(df[column].mean()[0])

        except Exception:
            df[column] = df[column].fillna(df[column].mode()[0])
        if column == 'WindGustDir' or column == 'WindDir9am':
            df[column] = df[column].fillna(df[column].mode()[0])
```

```python
df.shape
```

```python
# check NaN values
nan_counts = df.isna().sum()
nan_counts
```

```python
# Set the outlier detection threshold
threshold = 5
# iterate over each column in the df
for column in df.columns:
    if df[column].dtype == np.float64:
        # calculate the quartiles (Q1 and Q3)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)

        # calculate the IQR and outlier range
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        # identify the rows containing outliers

        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        # drop the outliers from the df
        df.drop(outliers.index, inplace=True)
```

```python
# for q3

def calculate_updated_mean(df, feature_column, new_data):

    existing_count = df[feature_column].count()
    existing_mean = df[feature_column].mean()
    existing_sum = existing_mean * existing_count
    new_sum = existing_sum + new_data[feature_column].count()
    updated_count = existing_count + len(new_data)

    updated_mean = new_sum / updated_count

    return updated_mean
```

```python
# for q3
def calculate_updated_count(df, feature_column, new_data):

    existing_count = df[feature_column].count()
    updated_count = existing_count + new_data[feature_column].count()

    return updated_count
```

```python
# for q3
def calculate_updated_max(df, feature_column, new_data):

    existing_max = df[feature_column].max()
    new_max = new_data[feature_column].max()
    updated_max = max(existing_max, new_max)

    return updated_max
```
